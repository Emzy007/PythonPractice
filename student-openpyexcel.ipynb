```python
import openpyxl as xl
import os
```

```python
os.getcwd()
```

```python
# Create a new workbook
student_wb = xl.Workbook()
# Create the first sheet
ws1 = student_wb.active
# By default the workbook will have 1 worksheet so renaming the default worksheet
```

```python
#Save the Excel workbook
wbname = input("Enter Excel file name in file.xlsx format: ")
if os.path.exists(os.getcwd()+f'/{wbname}'):
    print(f'{wbname} file exists! Enter a new name.')
else:
    student_wb.save(wbname)
    print(f'{wbname} Saved successfully!')

try:
    student_wb = xl.load_workbook(wbname)
    print(f'{wbname} loaded successfully!')
except:
    print(f'Unable to load {wbname} file')
#print(os.getcwd()+f'/{wbname}')
```
