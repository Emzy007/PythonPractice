```python
import openpyxl as xl
import os
```

```python
os.getcwd()
```

```python
# Create a new workbook
student_wb = xl.Workbook()
# Create the first sheet
ws1 = student_wb.active
# By default the workbook will have 1 worksheet so renaming the default worksheet
ws1.title = 'Sheet1'
print('List of worksheets: ', student_wb.sheetnames)
print('Active worksheet: ', student_wb.active)
```

```python
#Save the Excel workbook.
#The excel file name can be taken as user input by un-commenting the below line
#wbname = input("Enter Excel file name in file.xlsx format: ") 
wbname = 'student.xlsx'
if os.path.exists(os.getcwd()+f'{wbname}'):
    print(f'{wbname} file exists! Trying to load the existing {wbname} file if exists.')
else:
    student_wb.save(wbname)
    print(f'{wbname} Saved successfully!')

#Load the excel
try:
    student_wb = xl.load_workbook(wbname)
    print(f'{wbname} loaded successfully!')
except:
    print(f'Unable to load {wbname} file')
```

```python
#Print the entire sheet in tabular format
def print_sheet(worksheet_obj):
    all_rows = tuple(worksheet_obj.rows)
    for row in all_rows:
        for col in row:
            print(f'{col.value}',end='\t')
        print()
print_sheet(ws1)
```

```python
#Data entry
def data_entry():
    ws1.append(['Student_Id','Student_Name','Mark1','Mark2','Mark3'])
    ws1.append([101,'David',45,56,67])
    ws1.append([102,'John',67,76,63])
    ws1.append([103,'Mark',84,82,93])
    ws1.append([104,'Andrew',94,59,69])
    student_wb.save(wbname)
data_entry()
    
```

```python
def total_marks():
    #Create a new column
    ws1['F1'] = 'Total_Mark'
    #Calculate the total marks as calculated marks = sum(mark1 + mark2 + mark3)  * 0.6
    all_rows = tuple(ws1.rows)
    for i in range(2, ws1.max_row+1):
        ws1[f'F{i}'] = (ws1[f'C{i}'].value + ws1[f'D{i}'].value + ws1[f'E{i}'].value) * 0.6
    student_wb.save(wbname)
total_marks()
```
